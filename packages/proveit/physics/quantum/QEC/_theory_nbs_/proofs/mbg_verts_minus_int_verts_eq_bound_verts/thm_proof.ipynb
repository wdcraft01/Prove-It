#### (1) For some instantiations, we need to know that ZErrorChains$(K)$ are a special subset of errs$(K)$

#### (2) We need that $\text{int\_verts}(G) \subseteq \text{Vertices}(G)$ and $\text{bound\_verts}(G) \subseteq \text{Vertices}(G)$

```python
from proveit.physics.quantum.QEC import interior_vertices_subset_eq_of_all_vertices
interior_vertices_subset_eq_of_all_vertices
```

```python
interior_vertices_subset_eq_of_all_vertices.instantiate(num_forall_eliminations=3)
```

```python
# for convenience, label the interior vertices
interior_vertices = interior_vertices_subset_eq_of_all_vertices.instance_expr.instance_expr.instance_expr.lhs
```

```python
from proveit.physics.quantum.QEC import boundary_vertices_subset_eq_of_all_vertices
boundary_vertices_subset_eq_of_all_vertices
```

```python
boundary_vertices_subset_eq_of_all_vertices.instantiate(num_forall_eliminations=3)
```

```python
# for convenience, label the boundary vertices
boundary_vertices = boundary_vertices_subset_eq_of_all_vertices.instance_expr.instance_expr.instance_expr.lhs
```

```python
# And label the entire set of vertices
all_vertices = boundary_vertices_subset_eq_of_all_vertices.instance_expr.instance_expr.instance_expr.rhs
```

#### (3) We also need a generic instantiation of the XOR claim: a vertex $v$ is either an interior vertex or a boundary vertex but not both.

```python
from proveit.physics.quantum.QEC import error_vertex_is_interior_xor_boundary
error_vertex_is_interior_xor_boundary
```

```python
error_vertex_is_interior_xor_boundary.instantiate(num_forall_eliminations=3)
```

#### (4) Now we bring these elements together using the `bipartition_membership_eq_set_differences` theorem:

```python
from proveit.logic.sets import bipartition_membership_eq_set_differences
bipartition_membership_eq_set_differences
```

```python
from proveit import A, B, S
bipartition_membership_eq_set_differences_inst = bipartition_membership_eq_set_differences.instantiate(
        {A:interior_vertices, B:boundary_vertices, S:all_vertices}
)
```
