Statement on the graph theory side, involving cycle space and cycles

If all vertices are even degree, we have member of the cycle space.

The cycle space has a basis, the set of all face-cycles (all 3- and 4-edges long).

If all vertices of H are even degree, there exists a set of face cycles F1, F2, … Fn, of supergraph G (in which the subgraph H exists) such that
H = F1 ∆ F2 ∆ … ∆ Fn
(where we have defined the ∆ operation as the disjunctive union of the edges of graphs).

Homomorphism

Face cycles === Z plaquette operators, each of which is a stabilizer generator.
