```python
conc_02 = conc_01.inner_expr().rhs.operands[0].substitute(edges_def_inst_rev)
```

#### Effect a substitution for $\text{Graph}(V,E)$ in `conc_02`

```python
graph_v_e_eq_G = Equals(Graph(V, E), G).prove()
```

```python
conc_03 = graph_v_e_eq_G.sub_right_side_into(conc_02)
```

#### Eliminate our Skolem constants

```python
from proveit.logic import Exists
conc_04 = Exists.eliminate((V, E), conc_03, auto_simplify=False)
```

```python
conc_04.derive_right_via_equality(assumptions=[conc_04.lhs])
```
