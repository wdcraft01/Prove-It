Theorems (or conjectures) for the theory of <a class="ProveItLink" href="theory.ipynb">proveit.abstract_algebra.fields</a>
========

```python
import proveit
# Prepare this notebook for defining the theorems of a theory:
%theorems_notebook # Keep this at the top following 'import proveit'.
from proveit import k, p
from proveit.logic import Forall, InClass
from proveit.numbers import two, Add, Complex, Exp, Mult, NaturalPos, Prime, Rational, Real
from proveit.abstract_algebra import Fields, FiniteField
```

```python
%begin theorems
```

```python
rational_field = InClass(Rational, Fields(Add._operator_, Mult._operator_))
```

```python
real_field = InClass(Real, Fields(Add._operator_, Mult._operator_))
```

```python
complex_field = InClass(Complex, Fields(Add._operator_, Mult._operator_))
```

```python
# a place-holder until we can be more accurate
two_elem_finite_field = InClass(FiniteField(two), Fields(Add._operator_, Mult._operator_))
```

```python
# A more general theorem: 
# Forall primes p and forall natpos k, FiniteField(p^{k}) will be a field over the
# appropriate modular addition and multiplication
finite_fields_order_p_k = Forall(
        (p, k), InClass(FiniteField(Exp(p, k)),
                        Fields(Add._operator_, Mult._operator_)),
        domains = [Prime, NaturalPos]
)
```
