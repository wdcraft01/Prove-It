#### Related Theorems and Theorem Usage

```python
from proveit.linear_algebra.distance import manhattan_distance_2d
manhattan_distance_2d
```

```python
from proveit import a, b, c, d
from proveit.numbers import Real
manhattan_distance_2d.instantiate(
    assumptions=[InSet(a, Real), InSet(b, Real), InSet(c, Real), InSet(d, Real)])
```

```python
from proveit.numbers import one, two, three, four
manhattan_distance_2d.instantiate({a:one, b:two, c:three, d:four})
```

```python
from proveit.linear_algebra.distance import euclidean_distance_2d
euclidean_distance_2d
```

```python
euclidean_distance_2d.instantiate(
    assumptions=[InSet(a, Real), InSet(b, Real), InSet(c, Real), InSet(d, Real)])
```

```python
euclidean_distance_2d.instantiate({a:one, b:two, c:three, d:four})
```

```python
from proveit.numbers import zero, three, four
euclidean_distance_2d.instantiate({a:three, b:four, c:zero, d:zero})
```
