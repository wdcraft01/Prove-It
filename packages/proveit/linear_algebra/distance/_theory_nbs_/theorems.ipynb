Manhattan distance in 2D

```python
from proveit import x, y, X, ExprTuple
from proveit.logic import Equals, Forall
from proveit.numbers import Abs, Add, subtract
from proveit.linear_algebra import x_1, x_2, y_1, y_2
manhattan_distance_2d = Forall((x, y),
       Equals(ManhattanDistance(x, y),
             Add(Abs(subtract(x_1, y_1)), Abs(subtract(x_2, y_2)))),
       conditions=[Equals(x, ExprTuple(x_1, x_2)), Equals(y, ExprTuple(y_1, y_2))])
```
