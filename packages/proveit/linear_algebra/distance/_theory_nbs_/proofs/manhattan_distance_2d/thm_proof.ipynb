```python
import proveit
theory = proveit.Theory() # the theorem's theory
from proveit import defaults, display_provers # useful imports
from proveit import a, b, n, p, u, v, ExprTuple
from proveit.logic import Equals
from proveit.numbers import one, two
from proveit.linear_algebra import a_1, a_2, b_1, b_2
```

```python
%proving manhattan_distance_2d
```

```python
defaults.assumptions = manhattan_distance_2d.all_conditions()
```

```python
from proveit.linear_algebra.distance import manhattan_distance_def
manhattan_distance_def
```

```python
manhattan_distance_def_inst = manhattan_distance_def.instantiate(
    {n:two}, num_forall_eliminations=2)
```

```python
from proveit.linear_algebra.distance import minkowski_distance_def
minkowski_distance_def
```

```python
minkowski_distance_def_inst = minkowski_distance_def.instantiate(
     {p:one, n:two, u:ExprTuple(a_1, a_2), v:ExprTuple(b_1, b_2),
      a:ExprTuple(a_1, a_2), b:ExprTuple(b_1, b_2)},
     num_forall_eliminations=4)
```

```python
minkowski_distance_def.instantiate({p:one, n:two}, num_forall_eliminations=2)
```

```python
minkowski_distance_def.instantiate(
    {p:one, n:two, u:u, v:v}, num_forall_eliminations=3)
```

```python
u_eq_a1_a2 = Equals(u, ExprTuple(a_1, a_2)).prove()
```

```python
v_eq_b1_b2 = Equals(v, ExprTuple(b_1, b_2)).prove()
```

```python
left_sub_conc = u_eq_a1_a2.sub_left_side_into(minkowski_distance_def_inst)
```

```python
v_eq_b1_b2.sub_left_side_into(left_sub_conc)
```
